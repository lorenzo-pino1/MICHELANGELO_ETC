# Introduction

This notebook calculates the signal-to-noise ratio achieved on a transiting planete with MICHELANGELO.

The calculation assumes that MICHELANGELO performs similar to CRIRES+ in terms of efficiency and of adaptive optics performance. The key differences are that:
- MICHELANGELO has a broader simultaneous wavelength coverage;
- CRIRES+ has a few very broad orders that are "sliced" by the detectors (hence: low efficiency in some settings towards the border of the orders). MICHELANGELO will have much lower losses due to the narrower orders.

We therefore run the CRIRES+ ETC, and extrapolate the results at the center of each order to determine the expected signal-to-noise ratio by MICHELANGELO. The CRIRES+ settings L3262 and L3302 sample the peak of the blaze function of CRIRES+, and we use those for the simulation.

```python
import numpy as np
import json
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as const
from scipy.ndimage import uniform_filter1d
import subprocess
from IPython.display import display, JSON
```

## Running the CRIRES+ ETC.

The routine etc_cli.py takes a .json file as input. This contains the CRIRES+ ETC parameters. It is run via:

> python etc_cli.py crires2 etc_input.json -o etc_output.json

In the current example, we are observing a system with the following parameters:
- Star: $T_{eff} = 7500 \mathrm{K}$, blackbody, K = 6.41;
- airmass 1.2, 0.5 moon illumination, 2.5 precipitable water vapor;
- L3262 setting (center of the CRIRES+ blaze function);
- DIT = 300s, NDIT = 1

```python
etc_run_name = 'nominal'

# Keys to modify in the json.
dict_modify_json = {
    'Kmag':6.41,
    'T_eff':7500.,
    'airmass':1.2,
    'moon_fli':0.5,
    'pwv':2.5,
    'DIT':300
}
```

```python
# Load the JSON input file
def load_json(filename):
    with open(filename, 'r') as file:
        return json.load(file)

# Save the modified JSON input file
def save_json(data, filename, output_dir = 'etc_input_files/'):
    with open(output_dir+filename, 'w') as file:
        json.dump(data, file, indent=4)

# Modify input parameters interactively
def modify_parameters(input_data, dict_modify_json):
    for key, value in dict_modify_json.items():
        if key == 'Kmag':
            input_data['target']['brightness']['mag'] = value
        if key == 'T_eff':
            input_data['target']['sed']['temperature'] = value
        if key == 'airmass':
            input_data['sky']['airmass'] = value
        if key == 'moon_fli':
            input_data['sky']['moon_fli'] = value
        if key == 'pwv':
            input_data['sky']['pwv'] = value
        if key == 'DIT':
            input_data['timesnr']['DET1.DIT'] = value
    return input_data

# Run etc_cli.py with the modified JSON input
def run_etc_cli(etc_script, json_file, output_file, instrument_name="crires2", input_dir = 'etc_input_files/', output_dir = 'output_CRIRES_ETC_command_line/'):
    
    command = ["python3", etc_script, instrument_name, input_dir+json_file, "-o", output_dir+output_file+'.json']
    result = subprocess.run(command, capture_output=True, text=True)
    print("Run {}. Parameters: {}".format(json_file, dict_modify_json))
    print("STDOUT:", result.stdout)
    print("STDERR:", result.stderr)


# Paths to files
template_json_file = "etc_input_template.json" # Do not modify
this_run_json_file = "{}.json".format(etc_run_name)
etc_script = "etc_cli.py"

# Load, modify, save, and run
input_data = load_json(template_json_file)
modified_input_data = modify_parameters(input_data, dict_modify_json)
save_json(modified_input_data, this_run_json_file)
run_etc_cli(etc_script, json_file, output_file=etc_run_name)

# All L-band modes
# Load, modify, save, and run
input_data = load_json(template_json_file)
for mode in ['L3426', 'L3412', 'L3377', 'L3340', 'L3302', 'L3262', 'L3244']:
    modified_input_data = modify_parameters(input_data, dict_modify_json)
    input_data['instrument']['settingkey'] = mode
    save_json(modified_input_data, this_run_json_file)
    run_etc_cli(etc_script, json_file, output_file=etc_run_name+'_' + mode)
```

```python
fig_obstarget = plt.figure()
ax_obstarget = fig_obstarget.add_subplot(111)

fig_throughput = plt.figure()
ax_throughput = fig_throughput.add_subplot(111)

fig_throughput_1 = plt.figure()
ax_throughput_1 = fig_throughput_1.add_subplot(111)

for mode, color in zip(['L3426', 'L3412', 'L3377', 'L3340', 'L3302', 'L3262', 'L3244'], ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6']):
    etc_results_file = 'output_CRIRES_ETC_command_line/{}'.format(etc_run_name+mode)
    with open('{}.json'.format(etc_results_file)) as f:
        etc_output = json.load(f)['data']
    
    n_observations = 1
    
    wavelength_array = []
    obstarget_array = []
    sky_array = []
    
    stellar_noise_array = []
    sky_noise_array = []
    ron_noise_array = []
    dark_noise_array = []
    total_noise_array = []

    throughput_atmosphere_array = []
    throughput_blaze_array = []
    throughput_detector_array = []
    throughput_enslittedenergy_array = []
    throughput_instrument_array = []
    throughput_telescope_array = []
    throughput_totalinclsky_array = []
    
    for order in etc_output['orders']:
        order_name = order['order']
        for detector in order['detectors']:
            detector_name = detector['name']
            # detector.keys() = dict_keys(['name', 'noise_components', 'plots', 'psf', 'saturation', 'wavelength'])
    
            wavelength = np.array(detector['wavelength'])*1.e10 # Angstroms
            wavelength_array.append(wavelength)
    
            obstarget = np.array(detector['noise_components']['target'])
            obstarget_array.append(obstarget)
    
            sky = np.array(detector['noise_components']['sky'])
            sky_array.append(sky)
            
            # All the noise components are the variances. You can compare with the online ETC (what goes in the square root for the noise are the variances).
            #
            # So in practice these are:
            # noise_components.noise_terms.target_term = noise_components.obstarget
            # noise_components.noise_terms.sky_term = sky
            # noise_components.noise_terms.ron_term = noise_components.noise_info.ron2 * noise_components.noise_info.nspat * noise_components.noise_info.nspec
            # noise_components.noise_terms.dark_term = noise_components.noise_info.nspat * noise_components.noise_info.dit * noise_components.noise_info.dark
            #
            
            stellar_noise_term = np.array(detector['noise_components']['noise_terms']['target_term'])
            sky_noise_term = np.array(detector['noise_components']['noise_terms']['sky_term'])
            ron_noise_term = np.array(detector['noise_components']['noise_terms']['ron_term'])
            dark_noise_term = np.array(detector['noise_components']['noise_terms']['dark_term'])
            
            stellar_noise_array.append(stellar_noise_term)
            sky_noise_array.append(sky_noise_term)
            ron_noise_array.append(ron_noise_term)
            dark_noise_array.append(dark_noise_term)
    
            total_noise = np.sqrt(stellar_noise_term + sky_noise_term + ron_noise_term + dark_noise_term)/np.sqrt(n_observations)
            total_noise_array.append(total_noise)

            throughput_atmosphere = np.array(detector['plots']['throughput']['atmosphere'])
            throughput_blaze = np.array(detector['plots']['throughput']['blaze'])
            throughput_detector = np.array(detector['plots']['throughput']['detector'])
            throughput_enslittedenergy = np.array(detector['plots']['throughput']['enslittedenergy'])
            throughput_instrument = np.array(detector['plots']['throughput']['instrument'])
            throughput_telescope = np.array(detector['plots']['throughput']['telescope'])
            throughput_totalinclsky = np.array(detector['plots']['throughput']['totalinclsky'])
            
            throughput_atmosphere_array.append(throughput_atmosphere)
            throughput_blaze_array.append(throughput_blaze)
            throughput_detector_array.append(throughput_detector)
            throughput_enslittedenergy_array.append(throughput_enslittedenergy)
            throughput_instrument_array.append(throughput_instrument)
            throughput_telescope_array.append(throughput_telescope)
            throughput_totalinclsky_array.append(throughput_totalinclsky)

    
    wavelength_array = np.array(wavelength_array)
    obstarget_array = np.array(obstarget_array)
    sky_array = np.array(sky_array)
    stellar_noise_array = np.array(stellar_noise_array)
    sky_noise_array = np.array(sky_noise_array)
    ron_noise_array = np.array(ron_noise_array)
    dark_noise_array = np.array(dark_noise_array)

    throughput_atmosphere_array = np.array(throughput_atmosphere_array)
    throughput_blaze_array = np.array(throughput_blaze_array)
    throughput_detector_array = np.array(throughput_detector_array)
    throughput_enslittedenergy_array = np.array(throughput_enslittedenergy_array)
    throughput_instrument_array = np.array(throughput_instrument_array)
    throughput_telescope_array = np.array(throughput_telescope_array)
    throughput_totalinclsky_array = np.array(throughput_totalinclsky_array)
    
    ax_obstarget.plot(np.ravel(wavelength_array), np.ravel(obstarget_array),color=color, linewidth=0.5, label='{}'.format(mode))
    ax_obstarget.legend()
    ax_obstarget.set_xlim(32000, 38000)

    if mode == 'L3426':
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_atmosphere_array), linewidth = 0.5, color = 'C5', label = 'atmosphere')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_blaze_array), color = 'C0', label = 'blaze')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_detector_array), color = 'C1', label = 'detector')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_enslittedenergy_array), color = 'C2', label = 'ens. energy')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_instrument_array), color = 'C3', label = 'instrument')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_telescope_array), color = 'C4', label = 'telescope')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_totalinclsky_array), 'k', label = 'total incl.sky')
    else:
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_atmosphere_array), linewidth = 0.5, color = 'C5')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_blaze_array), color = 'C0')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_detector_array), color = 'C1')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_enslittedenergy_array), color = 'C2')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_instrument_array), color = 'C3')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_telescope_array), color = 'C4')
        ax_throughput.plot(np.ravel(wavelength_array), np.ravel(throughput_totalinclsky_array), 'k')

    if mode == 'L3426':
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_blaze_array), color = 'C0', label = 'blaze')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_totalinclsky_array), 'k', label = 'total incl.sky')
        ax_throughput_1.plot( np.ravel(throughput_blaze_array*throughput_detector_array*throughput_enslittedenergy_array*throughput_instrument_array*throughput_telescope_array), 'C1')
    else:
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_atmosphere_array), linewidth = 0.5, color = 'C5')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_blaze_array), color = 'C0')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_detector_array), color = 'C1')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_enslittedenergy_array), color = 'C2')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_instrument_array), color = 'C3')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_telescope_array), color = 'C4')
        ax_throughput_1.plot(np.ravel(wavelength_array), np.ravel(throughput_totalinclsky_array), 'k')

    ax_throughput.legend()
    ax_throughput.set_xlim(32000, 38000)
```

```python
etc_results_file = 'output_CRIRES_ETC_command_line/{}'.format(etc_run_name)
with open('{}.json'.format(etc_results_file)) as f:
    etc_output = json.load(f)['data']

n_observations = 1

wavelength_array = []
obstarget_array = []
sky_array = []

stellar_noise_array = []
sky_noise_array = []
ron_noise_array = []
dark_noise_array = []
total_noise_array = []

for order in etc_output['orders']:
    order_name = order['order']
    for detector in order['detectors']:
        detector_name = detector['name']
        # detector.keys() = dict_keys(['name', 'noise_components', 'plots', 'psf', 'saturation', 'wavelength'])

        wavelength = np.array(detector['wavelength'])*1.e10 # Angstroms
        wavelength_array.append(wavelength)

        obstarget = np.array(detector['noise_components']['target'])
        obstarget_array.append(obstarget)

        sky = np.array(detector['noise_components']['sky'])
        sky_array.append(sky)
        
        # All the noise components are the variances. You can compare with the online ETC (what goes in the square root for the noise are the variances).
        #
        # So in practice these are:
        # noise_components.noise_terms.target_term = noise_components.obstarget
        # noise_components.noise_terms.sky_term = sky
        # noise_components.noise_terms.ron_term = noise_components.noise_info.ron2 * noise_components.noise_info.nspat * noise_components.noise_info.nspec
        # noise_components.noise_terms.dark_term = noise_components.noise_info.nspat * noise_components.noise_info.dit * noise_components.noise_info.dark
        #
        
        stellar_noise_term = np.array(detector['noise_components']['noise_terms']['target_term'])
        sky_noise_term = np.array(detector['noise_components']['noise_terms']['sky_term'])
        ron_noise_term = np.array(detector['noise_components']['noise_terms']['ron_term'])
        dark_noise_term = np.array(detector['noise_components']['noise_terms']['dark_term'])

        stellar_noise_array.append(stellar_noise_term)
        sky_noise_array.append(sky_noise_term)
        ron_noise_array.append(ron_noise_term)
        dark_noise_array.append(dark_noise_term)

        total_noise = np.sqrt(stellar_noise_term + sky_noise_term + ron_noise_term + dark_noise_term)/np.sqrt(n_observations)
        total_noise_array.append(total_noise)

wavelength_array = np.array(wavelength_array)
obstarget_array = np.array(obstarget_array)
sky_array = np.array(sky_array)
stellar_noise_array = np.array(stellar_noise_array)
sky_noise_array = np.array(sky_noise_array)
ron_noise_array = np.array(ron_noise_array)
dark_noise_array = np.array(dark_noise_array)
```
